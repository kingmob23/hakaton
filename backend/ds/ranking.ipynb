```python
import polars as pl
video_stat = pl.read_parquet("video_stat.parquet")
video_stat
```

```python
# subset
ranking_df = video_stat.select(pl.col("video_id", "v_year_views", "v_month_views", "v_week_views", "v_day_views", "v_likes", "v_dislikes", "v_avg_watchtime_1_day", "v_avg_watchtime_7_day", "v_avg_watchtime_30_day", "v_category_popularity_percent_7_days", "v_category_popularity_percent_30_days", "v_pub_datetime", "v_long_views_1_days", "v_long_views_7_days", "v_long_views_30_days"))
ranking_df
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤—Å–µ —Å—Ç–æ–ª–±—Ü—ã –∫ —Ç–∏–ø—É Float64
df_float = ranking_df.select([
    pl.col(col).cast(pl.Float64).alias(col) if col != "video_id" else pl.col(col) 
    for col in ranking_df.columns
])
df_float
```

```python
# min-max —Ñ–æ—Ä–º—É–ª–∞ –∏–∑ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç–∞
scaled = df_float.select(
    [
        pl.col("video_id"),
        *(
            (pl.col(col) - pl.col(col).min()) / (pl.col(col).max() - pl.col(col).min())
            for col in df_float.columns
            if col != "video_id"
        ),
    ]
)
scaled
```

```python
# –ó–∞–¥–∞–µ–º –≤–µ—Å–∞ –¥–ª—è –∫–∞–∂–¥–æ–π –º–µ—Ç—Ä–∏–∫–∏
weights = {
    'v_year_views': 0.3,
    'v_month_views': 0.2,
    'v_likes': 0.15,
    'v_week_views': 0.1,
    'v_day_views': 0.05,
    'v_avg_watchtime_1_day': 0.05,
    'v_avg_watchtime_7_day': 0.05,
    'v_avg_watchtime_30_day': 0.05,
    'v_category_popularity_percent_7_days': 0.025,
    'v_category_popularity_percent_30_days': 0.025,
    'v_long_views_1_days': 0.025,
    'v_long_views_7_days': 0.025,
    'v_long_views_30_days': 0.025,
    'v_pub_datetime': 0.02,
    'v_dislikes': -0.05
}
```

```python
# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∏—Ç–æ–≥–æ–≤—ã–π score
df_with_scores = scaled.with_columns([
    sum(pl.col(col) * weight for col, weight in weights.items()).alias("final_score")
])

# –†–µ–∑—É–ª—å—Ç–∞—Ç —Å –∏—Ç–æ–≥–æ–≤—ã–º–∏ score
df_with_scores
```

```python
df_with_scores.sort("final_score", descending=True).select(pl.col("video_id"), pl.col("final_score"))
```
